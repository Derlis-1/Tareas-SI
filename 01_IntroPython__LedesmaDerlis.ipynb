```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.describe()
```

```python

porcentaje_supervivencia = df['Survived'].mean() * 100
print(f"Porcentaje de pasajeros que sobrevivieron: {porcentaje_supervivencia:.2f}%")
```

```python

porcentaje_parch = df['Parch'].value_counts(normalize=True) * 100
print("Porcentaje de pasajeros segun Parch (Padres/Hijos a bordo):")
print(porcentaje_parch)

print("\n--------------------------------------------------\n")


porcentaje_pclass = df['Pclass'].value_counts(normalize=True) * 100
print("Porcentaje de pasajeros segun Pclass (Clase del boleto):")
print(porcentaje_pclass)
```

```python

conteo_genero = df['Sex'].value_counts()
print("Cantidad de pasajeros por genero:")
print(conteo_genero)
```
